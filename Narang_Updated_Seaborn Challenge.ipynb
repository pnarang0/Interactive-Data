```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('insurance.csv')
```

```python
data.head()
```

```python
sns.relplot(data = data, x = 'bmi', y = 'charges', col = 'sex')
```

```python
# as there is no scatter in data, so i am using same column in hue as well.
sns.relplot(data = data, x = 'bmi', y = 'charges', col = 'sex', size = 'region', hue = 'region' )
```

```python
# There is no smoker_id in the data so i am mapping this column's data
data['smoker'] = data['smoker'].map({'yes': 'smoker', 'no': 'non-smoker'})
```

```python
# applying on this
sns.relplot(data = data, x = 'bmi', y = 'charges', hue = 'smoker',  palette = {'smoker' : '#A933FF', 'non-smoker' : '#FF5433'})
```
